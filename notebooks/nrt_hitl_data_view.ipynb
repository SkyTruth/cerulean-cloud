```python
import geopandas as gpd
import pandas as pd
from shapely import wkt
import matplotlib.pyplot as plt
```

```python
geo_file = r"C:\Users\ebeva\SkyTruth\cv3\source potential\allHITL-1767888944318.csv"
df = pd.read_csv(geo_file)
df["st_astext"] = df["st_astext"].apply(wkt.loads)
hitl_gdf = gpd.GeoDataFrame(
    df,
    geometry="st_astext",
    crs="EPSG:4326",  # adjust if coordinates are not lon/lat
)
hitl_gdf["perim_length_ratio"] = hitl_gdf["perimeter"] / hitl_gdf["length"]
```

```python
len(hitl_gdf[hitl_gdf["hitl_cls"] == 8])
```

```python
true_vessel_gdf = hitl_gdf[hitl_gdf["hitl_cls"].isin([6, 7, 8])]
false_slick_gdf = hitl_gdf[hitl_gdf["hitl_cls"] == 1]
```

```python
len(true_vessel_gdf)
```

```python
# true_vessel_gdf['perim_length_ratio'].hist(bins=50)
# false_slick_gdf['perim_length_ratio'].hist(bins=50)

plt.figure(figsize=(8, 6))
plt.hist(
    true_vessel_gdf["perim_length_ratio"].dropna(),
    bins=50,
    alpha=0.6,
    label="true_vessel_gdf",
    density=True,
)
plt.hist(
    false_slick_gdf["perim_length_ratio"].dropna(),
    bins=50,
    alpha=0.6,
    label="false_slick_gdf",
    density=True,
)
plt.xlabel("perim_length_ratio")
plt.ylabel("Density")
plt.title("Perimeter / Length ratio: true_vessel_gdf vs false_slick_gdf")
plt.legend()
plt.grid(True)
plt.show()
```
